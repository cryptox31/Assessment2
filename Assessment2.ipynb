```python
import numpy as np 
import pandas as pd 
from itertools import islice
from subprocess import check_output
print(check_output(["ls", "."]).decode("utf8"))
```

```python
df_exec = pd.read_excel("Execution Extract.xlsx")

#removing unwanted character ('[',']') from PRICE column
df_exec['PRICE'] = df_exec['PRICE'].map(lambda x: x.lstrip('[').rstrip(']'))
df_exec.insert(6,'ORDER_REF',0)
#changing datatype of PRICE column to float64
df_exec[['PRICE']]= df_exec[['PRICE']].apply(pd.to_numeric)

df_exec
```

```python
df_order = pd.read_excel("Order Log.xlsx")
df_order
```

```python
for i,row in df_exec.iterrows():
    for j,row2 in df_order.iterrows():
        if row[2]==row2[3]:
            if row[4]==row2[10]:
                df_exec.loc[i,'ORDER_REF']=row2[4]
                break
df_exec
```

# Write dataframe to excel sheet

```python
df_exec.to_excel("Execution Extraction.xlsx",index=False)
```
